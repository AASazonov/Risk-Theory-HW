```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

# HW 6

```python
def f1(x):
    return np.exp(-x)

def f2(x):
    return 1-x+0.25*x**2

x = np.linspace(0, 2, 100)

sns.lineplot(x=x, y=f2(x), label='$1-d+0.25d^2$')
sns.lineplot(x=x, y=f1(x), label='$exp(-d)$')
plt.savefig('pics/hw6t4p1.pdf')
```

# HW 07

```python
def FX(x, h: float = 1):
    if x<0:
        return 0
    elif x<3/h:
        return 2/3
    return 1

def FY(y, h: float = 1):
    if y<0:
        return 0
    elif y<2/(3*h):
        return 1/3
    elif y<3/h:
        return 2/3
    return 1

x = np.linspace(-1, 4, 1000)
f = np.zeros(x.size)
for i in range(x.size):
    f[i] = FX(x[i])
sns.lineplot(x=x, y=f, label='$F_X$')

for i in range(x.size):
    f[i] = FY(x[i])
sns.lineplot(x=x, y=f, label='$F_Y$', linestyle="dashed")
plt.savefig('pics/hw7t1p1.pdf')
```
